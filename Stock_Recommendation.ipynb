```python
import nltk
from sklearn.metrics.pairwise import cosine_similarity
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer


nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess(text):
    tokens = word_tokenize(text.lower())
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word.isalpha() and word not in stop_words]
    return ' '.join(tokens)
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('nifty_500.csv')
df.head()
```

```python
filtered_df = df[['Company Name', 'Industry']]
filtered_df
```

```python
from sentence_transformers import SentenceTransformer

# Load a pre-trained sentence transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

```

```python
customer_interest = "electric vehicles, renewable energy"
customer_interest_preprocessed = preprocess(customer_interest)
print(customer_interest_preprocessed)
```

```python
filtered_df
```

```python
filtered_df['Industry'] = filtered_df['Industry'].apply(preprocess)
```

```python
filtered_df
```

```python
stock_embeddings = model.encode(filtered_df['Industry'].tolist())
```

```python
customer_interest = input("Enter your interest: ")
customer_interest_preprocessed = preprocess(customer_interest)
customer_embedding = model.encode([customer_interest_preprocessed])[0]
```

```python
print("Stock Embeddings:\n", stock_embeddings)
print("\nCustomer Embedding:\n", customer_embedding)
```

```python
print("Stock Descriptions and their Embeddings:")
for name, embedding in zip(filtered_df['Company Name'], stock_embeddings):
    print(f"{name}: {embedding}")

print("\nCustomer Interest Embedding:")
print(customer_embedding)
```

```python
similarities = cosine_similarity([customer_embedding], stock_embeddings)

# Get the indices of the top similar stocks
top_indices = np.argsort(similarities[0])[-3:][::-1]  # Top 3 recommendations

# Display the recommended stocks
recommended_stocks = filtered_df.iloc[top_indices]
print("Recommended stocks for you based on your interest:")
print(recommended_stocks[['Company Name']])
```
